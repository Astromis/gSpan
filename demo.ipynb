```python
from gSpan.gspan_mining.graph import Graph

def read_graphs_from_list( graph_list:list, is_undirected=False):
    graphs = []
    for g_idx, graph in enumerate(graph_list):
        tgraph = Graph(g_idx,
                    is_undirected=is_undirected,
                    eid_auto_increment=False)
        for n in graph.nodes():
            if str(graph.nodes[n]['pos']) == '':
                print(n, g_idx)
            tgraph.add_vertex(n, graph.nodes[n]['pos'])

        for i,e in enumerate(graph.edges()):
            if e[1] == e[0]:
                continue
            tgraph.add_edge(i, e[0], e[1],  '')
        graphs.append(tgraph)
    return graphs
```
